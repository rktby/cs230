```python
import numpy as np
import pandas as pd
import tensorflow as tf

tf.enable_eager_execution()
tf.set_random_seed(231)

import time
from matplotlib import pyplot as plt
```

```python
hparams = tf.contrib.training.HParams(
    batch_size=100,
    in_seq_len=61,
    input_dim=5,
    out_seq_len=61,
    output_dim=1,
    num_layers=2,
    neurons_unit=32,
    learning_rate = 0.01,
    lr_decay = 0.99,
    lambd = 1e-6,
    datagen = 'prodn',
    logs_path = '/tmp/tensorflow_logs')
```

### Define Data Generator

```python
# Load production dataset
if hparams.datagen == 'prodn':
    prodn = pd.read_pickle('../full_6d.pkl').values
    dataset = np.nan_to_num(prodn)[2:102]

# Create synthetic sine wave data
if hparams.datagen == 'sin':
    dataset = np.zeros((hparams.batch_size, 10 * hparams.in_seq_len))
    dataset[1:,0] = np.random.randn(hparams.batch_size-1)
    for i in range(dataset.shape[1] - 1):
        dataset[:,i+1] = dataset[:,i] + 2 * np.pi / hparams.in_seq_len
    dataset = np.sin(dataset) + 1

# Reshape data
end_pos = hparams.in_seq_len * (hparams.input_dim - 1) + 1

# x.shape = (n_obs, in_seq_len, input_dim)
x = np.array([dataset[:,pos:-end_pos+pos] for pos in range(0,end_pos,hparams.in_seq_len)])
x = np.rollaxis(x, 0, 3) / np.max(dataset)

y = dataset[:,end_pos:, np.newaxis] / np.max(dataset)

del(dataset)
print(x.shape, y.shape)
```

```python
dataset = tf.data.Dataset.from_tensor_slices((x.astype(np.float32), y.astype(np.float32)))
dataset = dataset.batch(hparams.batch_size, drop_remainder=True)
```

### Define Model

```python
class Model(tf.keras.Model):
    def __init__(self, output_dim, units, batch_size):
        super(Model, self).__init__()
        self.units = units
        self.output_dim = output_dim
        self.batch_sz = batch_size

        self.fc_in = tf.keras.layers.Dense(1, activation='relu', name='affine_in')
        
        if tf.test.is_gpu_available():
            self.gru = tf.keras.layers.CuDNNGRU(self.units, 
                                                return_sequences=True, 
                                                return_state=True, 
                                                recurrent_initializer='glorot_uniform',
                                                name='gru')
        else:
            self.gru = tf.keras.layers.GRU(self.units, 
                                                return_sequences=True, 
                                                return_state=True, 
                                                recurrent_activation='relu', 
                                                recurrent_initializer='glorot_uniform',
                                                name='gru')

        self.fc_out = tf.keras.layers.Dense(self.output_dim, activation='relu', name='affine_out')

    def call(self, x, hidden):
        
        x = self.fc_in(x)
        
        # output shape == (batch_size, max_length, hidden_size) 
        # states shape == (batch_size, hidden_size)
        # states variable to preserve the state of the model
        # this will be used to pass at every step to the model while training
        output, states = self.gru(x, initial_state=hidden)

        # reshaping the output so that we can pass it to the Dense layer
        # after reshaping the shape is (batch_size * max_length, hidden_size)
        # output = tf.reshape(output, (-1, output.shape[2]))

        # The dense layer will output predictions for every time_steps(max_length)
        # output shape after the dense layer == (max_length * batch_size, vocab_size)
        x = self.fc_out(output)

        return x, states
```

### Train Model

```python
def loss_function(real, preds, variables, lambd = 0.0001):
    # Calculate OLS loss
    mse_loss = tf.losses.mean_squared_error(labels=real, predictions=preds)

    # Add L2 Regularisation
    l2_loss = 0
    for var in model.variables:
        l2_loss += lambd * tf.reduce_sum(var ** 2)

    accuracy, loss = mse_loss, mse_loss + l2_loss
    return loss, accuracy

def train_model(model, optimizer, dataset, hparams, epochs = 10, verbose = True):
    
    training_run_start = time.time()
    stats = []
    
    for epoch in range(epochs):
        epoch_start = time.time()

        for (batch, (inp, targets)) in enumerate(dataset):
            # initializing the hidden state at the start of every batch
            hidden = model.reset_states()

            with tf.GradientTape() as tape:
                # feeding the hidden state back into the model
                predictions, hidden = model(inp, hidden)

                # calculate model loss
                loss, accuracy = loss_function(targets, predictions, model.variables)

            # calculate gradients and update parameters
            grads = tape.gradient(loss, model.variables)
            optimizer.apply_gradients(zip(grads, model.variables))

            if batch % 10 == 0 and verbose:
                print('Epoch {} Batch {} Loss {:.4f} Accuracy {:.4f}'.format(\
                        epoch+1, batch, loss * 100, accuracy * 100))

        if verbose:
            print('Time taken for 1 epoch {:.4f} sec\n'.format(time.time() - epoch_start))

        # apply learning rate decay
        optimizer._lr *= hparams.lr_decay
        stats.append([loss, accuracy, time.time() - epoch_start])
    
    return loss, accuracy, time.time() - training_run_start, np.array(stats)
```

```python
# Run a training batch
tf.set_random_seed(231) # Set seed

# Initialise model and optimiser
model = Model(hparams.output_dim, hparams.neurons_unit, hparams.batch_size)
optimizer = tf.train.AdamOptimizer(learning_rate = hparams.learning_rate)

# Start training run
loss, accuracy, run_time, stats = train_model(model, optimizer, dataset, hparams, verbose=False, epochs=200)

# Display results
model.summary()
print('Loss {:.4f} Accuracy {:.4f} Time {:.4f}'.format(loss * 100, accuracy * 100, run_time))
plt.figure()
plt.plot(stats[:,:2])
plt.show()
```

### Test Hyperparameters

```python
np.random.seed(231)
models = []

#for nodes in np.floor(2 ** np.arange(3,9)):
#for lr in np.random.uniform(-8, -1, 10):
for lr in range(-8, 0, 1):
    lr /= 2

    # Run a training batch
    tf.set_random_seed(231) # Set seed

    # Initialise model and optimiser
    model = Model(hparams.output_dim, hparams.neurons_unit, hparams.batch_size)
    optimizer = tf.train.AdamOptimizer(learning_rate = 10 ** lr)

    # Start training run
    loss, accuracy, run_time, stats = train_model(model, optimizer, dataset, hparams, epochs = 40, verbose=False)
    print('Learning Rate {:.4f} Loss {:.4f} Accuracy {:.4f} Time {:.1f}'.format(lr, loss*100, accuracy * 100, run_time))
    
    models.append(model)
```

```python
# empty string to store our results
TARGET_INDEX = 49
BURN_IN = 2 * hparams.in_seq_len

forecast = []

_, (inp, target) = enumerate(dataset).__next__()
input_eval = tf.reshape(inp[TARGET_INDEX,0],(1,1,5))

# hidden state shape == (batch_size, number of rnn units); here batch size == 1
hidden = [tf.zeros((1, hparams.neurons_unit))]

for i in range(700):
    predictions, hidden = model(input_eval, hidden)

    # We pass the predicted word as the next input to the model
    # along with the previous hidden state
    if i < BURN_IN:
        input_eval = inp[TARGET_INDEX,i]
    else:
        input_eval = inp[TARGET_INDEX,i,:-1]
        input_eval = tf.concat((input_eval, predictions[0,0]), 0)
    
    input_eval = tf.reshape(input_eval,(1,1,5))
    
    forecast.append(np.squeeze(predictions))

plt.figure(figsize=(16,12))
plt.plot(forecast)
plt.plot(np.squeeze(target[TARGET_INDEX]))
plt.show()
```
